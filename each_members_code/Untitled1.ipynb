```python
# Dependencies and Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file_basic_path="data_sets/"
newFile = pd.ExcelFile("data_sets/climate_change_data.xls")
climchang_df = pd.read_excel("data_sets/climate_change_data.xls", sheet_name="Data")
climchang_df.columns

co2_df=climchang_df[climchang_df['Series name']=='CO2 emissions per capita (metric tons)'].copy()
co2_df

co2_df.drop(['Country code', 'Series code','Series name','SCALE','Decimals',2009,2010,2011], axis=1, inplace = True)


co2_df

#swaping index and column values around
co2_df_transposed = co2_df.transpose().copy()
co2_df_transposed

co2_df_transposed.reset_index(drop = False, inplace = True)

#making row index 0 into column titles
co2_df_transposed.columns = co2_df_transposed.iloc[0,:]

co2_df_transposed

#removing country name column
co2_df_transposed = co2_df_transposed.drop(0)

#Changing name from country name to year

co2_df_transposed = co2_df_transposed.rename(columns = {"Country name":"Year"})

co2_df_transposed
```
